```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import data
```

```python
# Define the corners of the rectangular surface
corners = [
    (0.0047625, -0.03227401003, -6.009082121E-15),
    (0.0047625, -0.03227401003, 3.048E-4),
    (-0.0047625, -0.03227401003, -5.981326545E-15),
    (-0.0047625, -0.03227401003, 3.048E-4)
]

# Extract x, y, and z ranges
x_min = min(corner[0] for corner in corners)
x_max = max(corner[0] for corner in corners)
z_min = min(corner[2] for corner in corners)
z_max = max(corner[2] for corner in corners)
y_value = corners[0][1]  # y is constant for all points

```

```python
# Define the number of points along each axis
num_points_per_axis = 25

# Generate a uniform grid
grid_x = np.linspace(x_min, x_max, num_points_per_axis)
grid_z = np.linspace(z_min, z_max, num_points_per_axis)
grid_points = [(x, y_value, z) for x in grid_x for z in grid_z]

# Turn this into a dataframe of x, y, z as columns
df = pd.DataFrame(grid_points, columns=['x', 'y', 'z'])
df

# Count the total number of points
print(f"Total number of points: {len(df)}")

# Plot the data for visualization
fig, ax = plt.subplots()
ax.scatter(df['x'], df['z'])
ax.set_xlabel('X-axis')
ax.set_ylabel('Z-axis')

# Save the dataframe in the data dir
# df.to_csv(data.DIR_PATH / 'init_pts_bitter_coils.csv', index=False, header=False)
# Instead of a csv save as a txt
df.to_csv(data.DIR_PATH / 'init_pts_bitter_coils.txt', index=False, header=False, sep=' ', float_format='%.16f')
```

```python
num_points_per_axis = 5

# Calculate the aspect ratio between the z and y ranges
z_range = z_max - z_min
x_range = x_max - x_min
print(z_range, x_range)
aspect_ratio = x_range / z_range
# print(aspect_ratio)

# Adjust the number of points along each axis to maintain uniformitx
num_points_x = int(num_points_per_axis * aspect_ratio)
num_points_z = num_points_per_axis

# Generate a uniform grid
grid_x = np.linspace(x_min, x_max, num_points_x)
grid_z = np.linspace(z_min, z_max, num_points_z)
grid_points = [(x, y_value, z) for x in grid_x for z in grid_z]

# Turn this into a dataframe of x, y, z as columns
df = pd.DataFrame(grid_points, columns=['x', 'y', 'z'])
df

# Count the total number of points
print(f"Total number of points: {len(df)}")

# Plot the data for visualization
fig, ax = plt.subplots()
ax.scatter(df['x'], df['z'])
ax.set_xlabel('X-axis')
ax.set_ylabel('Z-axis')
# Set aspect 
# ax.set_aspect('equal')

# Save the dataframe in the data dir
df.to_csv(data.DIR_PATH / 'init_pts_bitter_coils_uniform.txt', index=False, header=False, sep=' ', float_format='%.16f')
```

```python
# Set a random seed for reproducibility
np.random.seed(41)

# Generate 100 random points
num_points = 400
random_x = np.random.uniform(x_min, x_max, num_points)
random_z = np.random.uniform(z_min, z_max, num_points)
random_points = [(x, y_value, z) for x, z in zip(random_x, random_z)]

# Turn this into a dataframe of x, y, z as columns
df = pd.DataFrame(random_points, columns=['x', 'y', 'z'])
df

# Plot the data for visualization
fig, ax = plt.subplots()
ax.scatter(df['x'], df['z'])
ax.set_xlabel('X-axis')
ax.set_ylabel('Z-axis')

# Save the dataframe in the data dir
df.to_csv(data.DIR_PATH / 'init_pts_bitter_coils.csv', index=False, header=False)
```
