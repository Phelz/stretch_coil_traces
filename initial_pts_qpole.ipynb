```python
# Define the corners of the rectangular surface
# ! THIS IS GROUND
# corners = [
#     (0.1034488454, 0.0703161279, 0.03236850579),
#     (0.1034488454, 0.0715861379, 0.03236850579),
#     (0.1034488454, 0.0715861379, -0.05763149421),
#     (0.1034488454, 0.0703161279, -0.05763149421) 
# ]

# Define the corners of the rectangular surface
corners = [
(0.1034488454, 0.0757665179, 0.03999450579),
(0.1034488454, 0.0745003179, 0.03999450579),
(0.1034488454, 0.0745003179, 0.1299945058),
(0.1034488454, 0.0757665179, 0.1299945058),
]

# Extract x, y, and z ranges
# x is constant for all points
x_value = corners[0][0]  # x is constant for all points

y_min = min(corner[1] for corner in corners)
y_max = max(corner[1] for corner in corners)
z_min = min(corner[2] for corner in corners)
z_max = max(corner[2] for corner in corners)
```

```python
num_points_per_axis = 3

# Calculate the aspect ratio between the z and y ranges
z_range = z_max - z_min
y_range = y_max - y_min
aspect_ratio = z_range / y_range

# Adjust the number of points along each axis to maintain uniformity
num_points_y = num_points_per_axis
num_points_z = int(num_points_per_axis * aspect_ratio)

# Generate a uniform grid
grid_y = np.linspace(y_min, y_max, num_points_y)
grid_z = np.linspace(z_min, z_max, num_points_z)
grid_points = [( x_value, y, z) for y in grid_y for z in grid_z]

# Turn this into a dataframe of x, y, z as columns
df = pd.DataFrame(grid_points, columns=['x', 'y', 'z'])
df

# Count the total number of points
print(f"Total number of points: {len(df)}")

# Plot the data for visualization
fig, ax = plt.subplots()
ax.scatter( df['z'], df['y'])
ax.set_xlabel('X-axis')
ax.set_ylabel('Z-axis')
# Set aspect 
ax.set_aspect('equal')

# Save the dataframe in the data dir
df.to_csv(data.DIR_PATH / 'init_pts_qpole_uniform.txt', index=False, header=False, sep=' ', float_format='%.16f')
```
